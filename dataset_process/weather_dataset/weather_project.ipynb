```python
import pandas as pd

df=pd.read_csv("beach-weather-stations-automated-sensors-1.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Air Temperature'].fillna(df['Air Temperature'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Wet Bulb Temperature'], inplace=True)

df.dropna(subset=['Measurement Timestamp'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Heading'], inplace=True)
```

```python
df.isnull().sum()
```

```python
mean_barometric_pressure = df['Barometric Pressure'].mean()

# Replace NaN values in the 'Barometric Pressure' column with the calculated mean
df['Barometric Pressure'].fillna(mean_barometric_pressure, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Rain Intensity'].fillna(df['Rain Intensity'].median(), inplace=True)
df['Total Rain'].fillna(df['Total Rain'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Precipitation Type'], inplace=True)
```

```python
df
```

```python
df.to_csv("beach_weather_station_no_missing_values.csv")
```
